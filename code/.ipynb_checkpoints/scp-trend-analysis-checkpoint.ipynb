```python
# Extract year, month, hour
df['year'] = pd.DatetimeIndex(df['inc_datetime']).year
df['month'] = pd.DatetimeIndex(df['inc_datetime']).month
df['hour'] = pd.DatetimeIndex(df['inc_datetime']).hour
df.head()
```

```python
# Remove 2025 as it does not have data for complete year to avoid bias
df = df[df['year'] != 2025]
df.shape
```

## Collision Count by Time

```python
#tot_coll_per_yr = df.groupby('year')['collision_type']
#count_per_yr = df.groupby('year').size.reset_index(name='year_count')
collision_per_yr = df.groupby('year')['collision_type'].count()
collision_per_yr
```

```python
plt.bar(collision_per_yr.index, collision_per_yr.values, label='Yearly Collisions')
plt.xlabel('Year')
plt.ylabel('Number of Collisions')
plt.title('Yearly Collision Count')
plt.show()
```

From 2004 to 2006, the number of collisions show an upward trend with an approximate 3000 collision increase in 2005 and peak in 2006 before decreasing steadily until 2012. From 2013 to 2015, it follows a similar pattern where there is an increase with a noticeable increase in 2014 and peak in 2015 before decreasing again. There is a noticeable decrease in 2020 before the numbers slight rise up again until 2022. After 2022, the collision count decreases again. The decrease in number of collisions in 2020 could be attributed to the global lockdown during COVID-19 as people were required to stay at home.

# Factors of Collision Analysis

## Conditions (weather, light condition, road condition)

### Data Visualization

```python
df_edit = df[df['weather'] != 'Unknown']
df_edit = df[df['light_cond'] != 'Unknown']
df_edit = df[df['road_cond'] != 'Unknown']
```

```python
ax = sns.countplot(data=df_edit, x='collision_type', hue='weather')
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1))
ax.set_title('Collision type by weather')
ax.figure.set_size_inches(10, 4)
plt.show()
```

```python
ax = sns.countplot(data=df_edit, x='collision_type', hue='light_cond')
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1))
ax.set_title('Collision type by light condition')
ax.figure.set_size_inches(10, 4)
plt.show()
```

```python
ax = sns.countplot(data=df_edit, x='collision_type', hue='road_cond')
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1))
ax.set_title('Collision type by road condition')
ax.figure.set_size_inches(10, 4)
plt.show()
```

## Behaviors (Inattention, under influence, speeding)

```python
plt.bar(x='collision_type', y=['inattention_ind', 'und_infl_ind', 'speeding_ind'])
```

```python
# sns.countplot(data=df_edit, x='speeding_ind', hue='inattention_ind')
# plt.show()
```

```python
# Understand how collision types differ between behaviors.
```

```python
# Proportion of collision types by speeding
prop_df = pd.crosstab(df_edit['speeding_ind'], df_edit['collision_type'], normalize='index')
prop_df.plot(kind='bar', stacked=True, figsize=(6,4))
plt.ylabel('Proportion')
plt.title('Collision type proportion by speeding')
plt.show()
```

```python
# Proportion of collision types by under influence
prop_df = pd.crosstab(df_edit['und_infl_ind'], df_edit['collision_type'], normalize='index')
prop_df.plot(kind='bar', stacked=True, figsize=(6,4))
plt.ylabel('Proportion')
plt.title('Collision type proportion by under influence')
plt.show()
```

```python
# Proportion of collision types by inattention
prop_df = pd.crosstab(df_edit['inattention_ind'], df_edit['collision_type'], normalize='index')
prop_df.plot(kind='bar', stacked=True, figsize=(6,4))
plt.ylabel('Proportion')
plt.title('Collision type proportion by inattention')
plt.show()
```
