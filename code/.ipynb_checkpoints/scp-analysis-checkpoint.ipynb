### Temporal Patterns

Q1: How many collisions occur per year, month and hour? Are certain times higher than the others?

```python
df['collision_type'].value_counts()
```

```python
# Extract year, month, hour
df['year'] = pd.DatetimeIndex(df['inc_datetime']).year
df['month'] = pd.DatetimeIndex(df['inc_datetime']).month
df['hour'] = pd.DatetimeIndex(df['inc_datetime']).hour
df.head()
```

```python
# Remove 2025 as it does not have data for complete year to avoid bias
df = df[df['year'] != 2025]
df.shape
```

```python
df['hour'].value_counts()
```

```python
df['month'].value_counts()
```

```python
df['year'].value_counts()
```

#### Collision Count by Year

```python
#tot_coll_per_yr = df.groupby('year')['collision_type']
#count_per_yr = df.groupby('year').size.reset_index(name='year_count')
collision_per_yr = df.groupby('year')['collision_type'].count()
collision_per_yr
```

```python
plt.bar(collision_per_yr.index, collision_per_yr.values, label='Yearly Collisions')
plt.xlabel('Year')
plt.ylabel('Number of Collisions')
plt.title('Yearly Collision Count')
plt.show()
```

From 2004 to 2006, the number of collisions show an upward trend with an approximate 3000 collision increase in 2005 and peak in 2006 before decreasing steadily until 2012. From 2013 to 2015, it follows a similar pattern where there is an increase with a noticeable increase in 2014 and peak in 2015 before decreasing again. There is a noticeable decrease in 2020 before the numbers slight rise up again until 2022. After 2022, the collision count decreases again. The decrease in number of collisions in 2020 could be attributed to the global lockdown during COVID-19 as people were required to stay at home. The lower numbers after the year 2020 may also be attributed to the expansion of the link light rail station connecting North Seattle to Downtown with stations opening in Northgate, Roosevelt and UDistrict, providing transportation alternatives to those who want to travel traffic-free.

#### Collision Count by Month

```python
collision_per_mth = df.groupby('month')['collision_type'].count()
collision_per_mth

plt.bar(collision_per_mth.index, collision_per_mth.values, label='Monthly Collisions', color='orange')
plt.xlabel('Month')
plt.ylabel('Number of Collisions')
plt.title('Monthly Collision Count')
plt.show()
```

Overall, the number of collisions per month throughout the year ranges from about 15000 to almost 20000. The number does not seem to have any obvious differences between months. The highest number of collisions happen in October that may be due to the reduced daylight and slippery roads from more rainfall as the city fully transitions into fall. The lowest of that occurs in February that may be due to people being less motivated to drive and drivers being more alert due to winter conditions. 

Looking from season to season, the season with the most number of collisions is fall (September to November), followed by summer (June to August), spring (March to May) and winter (December to February). The season with the least obvious changes between its months is summer while the most noticeable changes between its months is fall.

#### Collision Count by Hour

```python
collision_per_hr = df.groupby('hour')['collision_type'].count()
collision_per_hr

plt.bar(collision_per_hr.index, collision_per_hr.values, label='Hourly Collisions', color='pink')
plt.xlabel('Hour')
plt.ylabel('Number of Collisions')
plt.title('Hourly Collision Count')
plt.show()
```

Overall, the number of collisions decreases from 12am to 4am before increasing and reaching first peak at 8am. After, the number decreases until 10am before increasing again and reaching second peak at 5pm. From 5pm to 11pm, the number of collisions decreases.

The highest number of collisions seem to occur at the 12am. However, some of the time was not reported in the dataset and automatically sets the time to 12am for unreported times. The number of collisions at 1am to 5am is lower than other times as there is less traffic at night, with the lowest number of collisions seem to occur at 4am, when most people are asleep.. Then, it peaks at 8am as it is peak hour for people going to the office. The number rises at 12pm to its highest peak (excluding 12am) at 5pm as more people are on the road to go on about their business. 5pm is the nest rush hour as most people working in the office are heading back home.

# Factors of Collision Analysis

## Conditions (weather, light condition, road condition)

```python
df_edit = df[df['weather'] != 'Unknown']
df_edit = df[df['light_cond'] != 'Unknown']
df_edit = df[df['road_cond'] != 'Unknown']
```

```python
ax = sns.countplot(data=df_edit, x='collision_type', hue='weather')
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1))
ax.set_title('Collision type by weather')
ax.figure.set_size_inches(10, 4)
plt.show()
```

```python
ax = sns.countplot(data=df_edit, x='collision_type', hue='light_cond')
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1))
ax.set_title('Collision type by light condition')
ax.figure.set_size_inches(10, 4)
plt.show()
```

```python
ax = sns.countplot(data=df_edit, x='collision_type', hue='road_cond')
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1))
ax.set_title('Collision type by road condition')
ax.figure.set_size_inches(10, 4)
plt.show()
```

## Behaviors (Inattention, under influence, speeding)

```python
plt.bar(x='collision_type', y=['inattention_ind', 'und_infl_ind', 'speeding_ind'])
```

```python
# sns.countplot(data=df_edit, x='speeding_ind', hue='inattention_ind')
# plt.show()
```

```python
# Understand how collision types differ between behaviors.
```

```python
# Proportion of collision types by speeding
prop_df = pd.crosstab(df_edit['speeding_ind'], df_edit['collision_type'], normalize='index')
prop_df.plot(kind='bar', stacked=True, figsize=(6,4))
plt.ylabel('Proportion')
plt.title('Collision type proportion by speeding')
plt.show()
```

```python
# Proportion of collision types by under influence
prop_df = pd.crosstab(df_edit['und_infl_ind'], df_edit['collision_type'], normalize='index')
prop_df.plot(kind='bar', stacked=True, figsize=(6,4))
plt.ylabel('Proportion')
plt.title('Collision type proportion by under influence')
plt.show()
```

```python
# Proportion of collision types by inattention
prop_df = pd.crosstab(df_edit['inattention_ind'], df_edit['collision_type'], normalize='index')
prop_df.plot(kind='bar', stacked=True, figsize=(6,4))
plt.ylabel('Proportion')
plt.title('Collision type proportion by inattention')
plt.show()
```
